```python
from control import margin, TransferFunction 
from control.matlab import bode
import scipy.signal as sig
import warnings
from matplotlib.pyplot import figure, semilogx, show, hold
```

## Specifications in the s-plane

```python
zeros = [0, -6]
poles = [-1, -3, -4, -4]
gain = 1

den = sig.TransferFunction(sig.ZerosPolesGain(zeros, poles, gain)).den 
num = sig.TransferFunction(sig.ZerosPolesGain(zeros, poles, gain)).num 

tf = TransferFunction(num, den)
```

```python
warnings.filterwarnings('ignore')
bode(tf)
warnings.filterwarnings('default')
```

```python
display(tf.to_ss())
```

```python
display(margin(tf.to_ss()))
```
