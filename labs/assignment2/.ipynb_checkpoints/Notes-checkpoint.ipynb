```python
from control import TransferFunction, bode
from control.matlab import damp, margin
import warnings
from numpy import sin
```

```python
# Prerequisites
n = 10 
Jt = 9.87 * 10**(-6) 
L = 1.7 * 10**(-4) 
Rl = 4 Kt = 2.13 * 10 **(-2) R = 2.14
tr = 36 * 10**(-3)
beta = Rl / n
```

```python
num = [ beta * Kt ]
den = [ L * Jt,  (R + Rl)* Jt, Kt ** 2]
```

```python
tf = TransferFunction(num, den)
display(tf)
warnings.filterwarnings("ignore")
bode(tf)
warnings.filterwarnings("default")
```

```python
wBW = 1.8 / tr
damp(tf, True)
```
